### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

#### Seaborn is a Python data visualization library built on top of Matplotlib that provides a high-level interface for creating informative and attractive statistical graphics. There are numerous types of plots you can create using Seaborn, each with its own specific use case. Here are five common types of plots and their uses:

1. Scatter Plot (sns.scatterplot):Use: Scatter plots are used to visualize the relationship between two continuous variables. They are helpful for identifying patterns, correlations, clusters, or outliers in the data.

2. Line Plot (sns.lineplot):Use: Line plots are used to display the trend or change in one or more variables over a continuous range. They are particularly useful for showing data over time or across ordered categories.

3. Bar Plot (sns.barplot):Use: Bar plots are used to compare categories of data by displaying the heights or lengths of bars. They are effective for showing the distribution of a categorical variable or comparing different categories.

4. Histogram (sns.histplot):Use: Histograms are used to visualize the distribution of a single continuous variable. They group data into bins and display the frequency or density of data points in each bin, making it easy to identify the data's underlying distribution.

5. Box Plot (sns.boxplot):Use: Box plots, also known as box-and-whisker plots, are used to display the distribution and spread of a continuous variable or multiple variables across different categories. They provide information about central tendency, variability, and potential outliers in the data.


#### These are just a few examples of the plots you can create with Seaborn. Seaborn also provides many other types of plots, including violin plots, swarm plots, heatmaps, pair plots, and more, each with its own specific use case for data exploration and visualization. The choice of which plot to use depends on the nature of your data and the insights you want to convey.


### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

Note:  timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot using Seaborn
plt.figure(figsize=(10, 6))  # Set the figure size (optional)

# Use Seaborn's lineplot function to create the plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

# Add labels and title
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Signal over Time by Event and Region")

# Show the plot
plt.legend(title="Event")
plt.show()
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note:  pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Create a figure with two subplots
plt.figure(figsize=(12, 5))

# Create the first box plot for 'age'
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Box Plot of Age by Pclass')

# Create the second box plot for 'fare'
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Box Plot of Fare by Pclass')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset. 

```python
diamond_data = sns.load_dataset("diamonds")
```

```python
diamond_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Create a histogram with hue for the 'cut' column
plt.figure(figsize=(10, 6))
sns.histplot(data=diamond_data, x='price', hue='cut', multiple='stack', bins=30)

# Add labels and title
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Histogram of Diamond Prices by Cut')

# Show the plot
plt.legend(title='Cut')
plt.show()
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
Flights_data = sns.load_dataset("flights")
```

```python
Flights_data
```

```python
# Pivot the dataset to create a matrix suitable for a heatmap
Flights_pivot = Flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(Flights_pivot, cmap='coolwarm', annot=True, fmt='d', linewidths=.5)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Month')
plt.title('Passenger Count by Month and Year')

# Show the plot
plt.show()
```
